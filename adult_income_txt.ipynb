## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_train_pred=dtc.predict(X_train)
y_test_pred=dtc.predict(X_test)
```

```python
print("Train Data")
print(classification_report(y_train,y_train_pred))

print("Test Data")
print(classification_report(y_test,y_test_pred))
```

```python
print("Train Data")
print(accuracy_score(y_train,y_train_pred))
print("Test Data")
print(accuracy_score(y_test,y_test_pred))
```

Above accuracies shows our model is overfitted so we have to use pruning technique to solve this problem.

```python
max_depth = []
acc_gini = []
acc_entropy = []
for i in range(1,21):
 gdtree = DecisionTreeClassifier(criterion='gini', max_depth=i)
 gdtree.fit(X_train, y_train)
 pred = gdtree.predict(X_test)
 acc_gini.append(accuracy_score(y_test, pred))
 ####
 edtree = DecisionTreeClassifier(criterion='entropy', max_depth=i)
 edtree.fit(X_train, y_train)
 pred = edtree.predict(X_test)
 acc_entropy.append(accuracy_score(y_test, pred))
 ####
 max_depth.append(i)
```

```python
d = pd.DataFrame({'acc_gini':pd.Series(acc_gini),
 'acc_entropy':pd.Series(acc_entropy),
 'max_depth':pd.Series(max_depth)})
```

```python
# visualizing changes in parameters
plt.plot('max_depth','acc_gini', data=d, label='gini')
plt.plot('max_depth','acc_entropy', data=d, label='entropy')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend()
```

```python
dtree = DecisionTreeClassifier(criterion='entropy', max_depth=8)
dtree.fit(X_train, y_train)
y_test_pred = dtree.predict(X_test)
y_train_pred = dtree.predict(X_train)
print('Train_dtree: ',accuracy_score(y_train, y_train_pred))
print('Test_dtree: ',accuracy_score(y_test, y_test_pred))
```

Decision Tree before Pruning

test accuracy: 78.99%
Decision Tree After Pruning
train accuracy: 82.70%
test accuracy: 82.51%

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 10, criterion = 'entropy')
rfc.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
def metric(y_act,y_pred):
    print(classification_report(y_act,y_pred))
    print(accuracy_score(y_act,y_pred))
    print(confusion_matrix(y_act,y_pred))
```

```python
print('Train data')
print(metric(y_train,y_train_pred))
print('Test data')
print(metric(y_test,y_test_pred))
```

```python
rfc_param_grid={'max_depth':[3,5,7,10],'n_estimators':[50,100,150],'criterion':['gini','entropy']}
```

```python
from sklearn.model_selection import GridSearchCV
grid_rfc= GridSearchCV(rfc,param_grid=rfc_param_grid,scoring='f1_macro',cv=5,n_jobs=-1)
```
